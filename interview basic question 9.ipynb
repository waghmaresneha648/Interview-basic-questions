```python
import pandas as pd
import numpy as np
df=pd.DataFrame({'User_ID': [101, 102, 103, 104],
    'City': ['New York', 'Paris', 'Tokyo', 'London'],
    'Device_Type': ['Mobile', 'Desktop', 'Mobile', 'Mobile'],
    'Actions': [5, 12, 3, 8]
                })
df['City'] = df['City'].astype('category')
df['Device_Type'] = df['Device_Type'].astype('category')

# The Naive Groupby
naive_result = df.groupby(['City', 'Device_Type'])['Actions'].sum()

print("Naive Groupby Output (Notice the 0.0 values for non-existent combinations):")
print(naive_result)
```
