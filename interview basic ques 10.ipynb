```python
import pandas as pd
import numpy as np 
df= pd.DataFrame({
    'Appilcation_id':[1,2,3],
    'credit_score':[580,780,650],
    'DTI_Ratio':[50,20,35]
})
conditions=[
            (df['credit_score'] < 600) & (df['DTI_Ratio'] > 45),
            (df['credit_score'] > 750)
           ]
choices=['High Risk','Low Risk']
df['Risk_Level']=np.select(conditions,choices,default='Medium Risk')
print(df)
```
