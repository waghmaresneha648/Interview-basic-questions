```python

default_profile = {"status": "Active", "scores": []}

# THE TRAP APPROACH 

users_group_A = [default_profile] * 3

# Let's add a score of 95 to the FIRST user only
users_group_A[0]["scores"].append(95)

print("Group A (The Trap):")
print(users_group_A)



# THE CORRECT APPROACH 

# Using a list comprehension creates a brand-new object on each iteration.
users_group_B = [{"status": "Active", "scores": []} for _ in range(3)]



# Now let's add a score to the FIRST user only
users_group_B[0]["scores"].append(95)

print("\nGroup B (The Correct Way):")
print(users_group_B)
```
