```python
#How do you unpack these comma-separated strings so that you can count every individual instance of the word 'Bug' across the whole dataset
```

```python
import pandas as pd

df = pd.DataFrame({
    'Ticket_ID': [1, 2, 3, 4],
    'Tags': ['Bug, Login', 'Billing', 'UI, Bug, Crash', 'Login, UI']
})

# 1. Split the string into an actual Python list of items
df['Tag_List'] = df['Tags'].str.split(', ')

# 2. "Explode" the list into individual rows
exploded_df = df.explode('Tag_List')

print("The Exploded DataFrame:")
print(exploded_df[['Ticket_ID', 'Tag_List']])

# 3. Now get the true individual counts
print("\nTrue Tag Counts:")
print(exploded_df['Tag_List'].value_counts())
```
