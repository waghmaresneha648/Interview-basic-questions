```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Transaction_ID': [1, 2, 3, 4, 5],
    'Amount': [100, -50, np.nan, -20, -40]
})

# Running the naive cumulative sum
df['Balance'] = df['Amount'].cumsum()

print(df)
```
