```python
#Combined table showing each premium user and how much they paid this month so they can calculate total revenue.
```

```python
import pandas as pd

users_df = pd.DataFrame({
    'User_ID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie']
})

payments_df = pd.DataFrame({
    'User_ID': [101, 102, 102, 103],
    'Amount': [15, 15, 15, 15]
})

print("Original Users Rows:", users_df.shape[0])

# Performing the merge
merged_df = pd.merge(users_df, payments_df, on='User_ID')

print("Merged DataFrame Rows:", merged_df.shape[0])
print("\nThe Merged Data:")
print(merged_df)
```
