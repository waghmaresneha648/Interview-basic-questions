```python
import pandas as pd

df = pd.DataFrame({
    'Date': pd.to_datetime(['2026-01-01', '2026-01-03', '2026-01-06', '2026-01-07']),
    'Revenue': [150, 200, 100, 300]
})

# 1. Date operations require the date column to be the Index
df = df.set_index('Date')


# 3. Use .fillna(0) because a day with no sales means $0 revenue
daily_report = df.resample('D').asfreq().fillna(0)

print(daily_report)
```
