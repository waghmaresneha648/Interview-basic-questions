```python
#You are an analyst at an e-commerce company. The marketing team ran a massive campaign last month. The CMO asks you to calculate the average order value (AOV) for the month to see if the campaign attracted big spenders.
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'order_id':[1001,1002,1003,1004],
    'order_amount':[50,100,np.nan,40]
})
print(df)
    
```

```python
print(df['order_amount'].mean())
```

```python
true_values=(df['order_amount'].fillna(0).median()
           )
print(true_values)
```

```python

print("correct amount is :",true_values)
```
